```python
import pandas as pd
```

```python
#This first csv contains way more data columns related to diabetes, life expectancy etc.than needed and is super huge; the one that is uncommented contains very precise data
#df = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
df = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/full_data.csv")
countries = pd.read_csv("countries.csv")
```

```python
a = countries['name'].unique().tolist()
b = df['location'].unique().tolist()


exclude_list = [i for i in b if i not in a]

#only including the countries for which we have the lat and long for
for i in exclude_list:
    df = df[df['location'] != i]
```

```python
#getting the lat and long for the countries
lat = [countries[countries['name'] == i]['latitude'].item() for i in df['location']]
long = [countries[countries['name'] == i]['longitude'].item() for i in df['location']]
```

```python
#accordingly appending it in our df 
df['Long'] = long
df['Lat'] = lat
```
